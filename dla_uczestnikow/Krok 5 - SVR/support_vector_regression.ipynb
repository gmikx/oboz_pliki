# Support Vector Regression (SVR)

## Importowanie bibliotek

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importowanie Zbioru Danych

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
y = y.reshape(len(y),1)
```

```python
print(y)
```

## Skalowanie Zmiennych

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

```python
print(X)
```

```python
print(y)
```

## Trenowanie SVR na Całym Zbiorze Danych

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X, y)
```

## Przewidywanie Wyników z Nowych Danych

```python
sc_y.inverse_transform([regressor.predict(sc_X.transform([[6.5]]))])[0]
```

## Wizualizacja Wyników SVR

```python
plt.scatter(sc_X.inverse_transform(X), sc_y.inverse_transform(y), color = 'red')
plt.plot(sc_X.inverse_transform(X), sc_y.inverse_transform([regressor.predict(X)])[0], color = 'blue')
plt.title('Truth or Bluff (SVR)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## Wizualizacja Wyników SVR (wyższa rozdzielczość i gładsza krzywa)

```python
X_grid = np.arange(min(sc_X.inverse_transform(X)), max(sc_X.inverse_transform(X)), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(sc_X.inverse_transform(X), sc_y.inverse_transform(y), color='red')
plt.plot(X_grid, sc_y.inverse_transform([regressor.predict(sc_X.transform(X_grid))])[0], color='blue')
plt.title('Position vs Salary (SVR)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```
