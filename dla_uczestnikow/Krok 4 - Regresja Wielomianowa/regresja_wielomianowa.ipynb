# Regresja Wielomianowa

## Importowanie bibliotek

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importowanie Zbioru Danych

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

## Trenowanie Regresji Liniowej na Całym Zbiorze Danych

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

## Trenowanie Regresji Wielomianowej na Całym Zbiorze Danych

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly, y)
```

## Wizualizacja Wyników Regresji Liniowej

```python
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg.predict(X), color='blue')
plt.title('Position vs Salary (Linear Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

## Wizualizacja Wyników Regresji Wielomianowej

```python
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg2.predict(X_poly), color='blue')
plt.title('Position vs Salary (Polynomial Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

## Wizualizacja Wyników Regresji Wielomianowej (wyższa rozdzielczość i gładsza krzywa)

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color='red')
plt.plot(X_grid, lin_reg2.predict(poly_reg.fit_transform(X_grid)), color='blue')
plt.title('Position vs Salary (Polynomial Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

## Przewidywanie Wyników z Nowych Danych - Regresja Liniowa

```python
lin_reg.predict([[6.5]])
```

## Przewidywanie Wyników z Nowych Danych - Regresja Wielomianowa

```python
lin_reg2.predict(poly_reg.fit_transform([[6.5]]))
```
